```python
# try_else문
try:
    age=input('나이를 입력하세요: ')
except:
    print('입력이 정확하지 않습니다.')
else:
    try:
        if age <= 18:
            print('미성년자는 출입금지입니다.')
        else:
            print('환영합니다.')
    except:
        print("프로그램 점검요망.")
```

```python
# process_scores.py
students = ["김철수", "이영희", "박민수", "최유진"]

for (1,student) in students:
    try:
        with open(f"{student}_성적.txt", 'r') as f:
            score = f.read()
            print(f"{student}의 성적: {score}")
    except FileNotFoundError:
        print(f"{student}의 성적 파일이 없습니다. 건너뜁니다.")
        continue # 다음 학생으로 넘어감
```
