## 여러 개의 입력값을 받는 함수 만들기

```python
def add_mul(choice,  *args): 
    if choice == "add":   # 매개변수 choice에 "add"를 입력받았을 때
        result = 0 
        for i in args: 
            result = result + i 
    elif choice == "mul":   # 매개변수 choice에 "mul"을 입력받았을 때
        result = 1 
        for i in args: 
            result = result * i 
    return result 
```

```python
result = add_mul('add', 1,2,3,4,5)
print(result)
result = add_mul('mul', 1,2,3,4,5)
print(result)
```

```python
# 위 예제에서 args를 하나만 넣고도 동작하도록 만들어보기
def add_mul(*args): 
    if choice == "add":   # 매개변수 choice에 "add"를 입력받았을 때
        result = 0 
        for i in args: 
            result = result + i 
    elif choice == "mul":   # 매개변수 choice에 "mul"을 입력받았을 때
        result = 1 
        for i in args: 
            result = result * i 
    return result 
```

## 함수의 반환값은 언제나 하나이다 *

```python
def add_and_mul(a,b): 
    # return a+b, a*b
    # return [a+b, a*b]
    return {a+b : a*b}
```

```python
value = add_and_mul(3,5)
print(value,type(value))
```

## 매개변수에 초깃값 미리 설정하기 *

```python
# 매개변수에 초깃값 미리 설정은 함수를 생성할 때 만든다.
def say_myself(name, age, man=True): 
    print("나의 이름은 %s 입니다." % name) 
    print("나이는 %d살입니다." % age) 
    if man: 
        print("남자입니다.")
    else: 
        print("여자입니다.")
say_myself("홍길동",22)
say_myself("박서방",23,man=True)
say_myself("향단이",24,man=False)
```

```python
# default2.py
def say_myself(name, man=True, age): 
    print("나의 이름은 %s 입니다." % name) 
    print("나이는 %d살입니다." % age) 
    if man: 
        print("남자입니다.") 
    else: 
        print("여자입니다.")

# 매개변수 초기화는 마지막에 둬야한다. 그뒤에 일반 매개변수가 오면 안된다.
```

## 변수의 효력 범위 *

```python
# 함수 안에서 사용하는 매개변수는 함수 밖의 변수 이름과는 전혀 상관없다.
a = 1
def vartest(a):
    a = a + 1
vartest(a) 
print(a)
```

```python
# vartest_return.py
a = 1 
def vartest(a): 
    a = a + 1 
    return a

a = vartest(a) 
print(a)
```

```python
# vartest_global.py
통장 = 1 
def 치트(): 
    global 통장 
    통장 = 통장 + 1

print(통장)
치트() 
print(통장)
```

```python
import requests as req
url = "https://finance.naver.com/sise/sise_market_sum.naver"
ori = req.get(url).text

code = ['005930','000270','005380','012450','068270']

def finance(code = '005930'):
    start_title = ori.find(f'<a href="/item/main.naver?code={code}" class="tltle">')
    ori1 = ori[start_title : start_title + 70].replace(f'<a href="/item/main.naver?code={code}" class="tltle">','').replace('</a></td>','').replace('</a></td','').replace('\n','').strip()
    ori2 = ori[start_title + 70 : start_title + 105].replace('<td class="number">','').replace('</td>','').replace('>','').strip()
    print("종목:",ori1)
    print("가격:",ori2)

for x in code:
    finance(x)
```
