### 리스트 관련 함수

````python
# append - 추가하기 *
a = []
print(a)
a.append(1)
print(a)
a.append(2)
print(a)
a.append([11,22,33])
print(a)
a.append(4)
print(a)
````

````python
# sort 리스트 정렬 *
a = [5,43,5,6,56,3,2,4,1]
b = ['gf','f','u','a','e','f','z','c']
a.sort()
b.sort()
print(a)
print(b)
````

````python
# reverse 리스트 뒤집기
a = [5,43,5,6,56,3,2,4,1]
b = ['gf','f','u','a','e','f','z','c']
print(a[::-1])
print(b[::-1])
a.reverse()
b.reverse()
print(a)
print(b)
````

````python
# sort 와 reverse 응용
a = [5,43,5,6,56,3,2,4,1]
a.sort()
a.reverse() #  a[::-1]
print(a)
b = [5,43,5,6,56,3,2,4,1]
b.sort(reverse=True) #True는 내림차순, False 오름차순이 기본값
print(b)
````

````python
# sorted 원본 유지하기
aa = [66,7,8,99,0]
print(sorted(aa))
print(sorted(aa,reverse=True))
print(aa)
````

````python
# sorted 응용하기
fruit = ['banana', 'kiwi', 'orange', 'melon', 'apple', 'strawberry']
print(sorted(fruit))
print(sorted(fruit,key=len)) # 글자 길이의 오름차순
print(sorted(fruit,key=len,reverse=True)) # 글자 길이의 내림차순
````

### 변수를 만드는 여러가지 방법

````python
a , b = 123, 456
print(a)
print(b)
````

````python
(a , b , c ) = (123 , 456, 789)
print(a)
print(b)
print(c)
````

````python
[a, b, c] = ['python','life','cool']
print(a)
print(b)
print(c)
````

````python
a = b = c = 0
print(a)
print(b)
print(c)
````

````python
# 두 변수의 값을 바꿀 때 (*)
a = 33
b = 55
print(a, b)

a, b = b, a
print(a, b)
````

````python
# 다른언어 예 temp
aa = 33
bb = 55
print(aa, bb)

temp = aa
aa = bb
bb = temp
print(aa, bb)
````
