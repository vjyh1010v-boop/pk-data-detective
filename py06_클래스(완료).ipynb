```python
# 예제1

class DDuk:
    def __init__(self, won, ea):
        self.won = won
        self.ea = ea

    def dduk(self):
        result = self.won * self.ea
        print(f"배가고파서 떡볶이집에서 떡볶이를 {result}원어치 먹었다.")

    def odeng(self):
        result = self.won * 2 * self.ea
        print(f"배가고파서 떡볶이집에서 오뎅을 {result}원어치 먹었다.")    
```

```python
재형 = DDuk(2500,4)
호성 = DDuk(1500,7)
재훈 = DDuk(2000,5)
```

```python
재형.dduk()
재형.odeng()
```

```python
호성.dduk()
호성.odeng()
```

```python
재훈.dduk()
재훈.odeng()
```

```python
# 예제2
class Pet:
    def __init__(self, name, species):
        """펫을 생성"""
        self.name = name
        self.species = species
        self.hunger = 50 # 배고픔 수치 0 ~ 100    
    def feed(self, many = 10):
        """밥을 주면 배고픔 감소"""
        if self.hunger > 10:
            self.hunger -= many
        return f"{self.name}에게 밥을 줬어요. 배고픔:{self.hunger}"
    def play(self, how = 20):
        """놀아주면 배고픔 증가"""
        if self.hunger < 80:
            self.hunger += how
        return f"{self.name}와 잘 놀았어요~!! 배고픔:{self.hunger}"
```

```python
dog1 = Pet("뽀삐","말티즈")
dog2 = Pet("강쥐","허숙히")
cat1 = Pet("컁이","도둑고양이")
cat2 = Pet("킁이","집고양이")
```

```python
cat1.feed(20)
```

```python
cat1.play(30)
```

```python
print(cat1.feed())
print(cat1.play())
```

```python
# 예제3
class Bacteria:
    def __init__(self, init_population, growth_rate):
        """세균초기화"""
        self.population = init_population
        self.growth = growth_rate
        self.time = 0 # 시작시간
    def grow(self, hours):
        """일정 시간이 지나면 세균이 증식"""
        self.time = hours
        self.population *= self.growth**hours
        return f"{self.time}시간이 흐릅니다. "
    def check(self):
        """현재 세균 수 확인"""
        print (f"현재 세균수 : {self.population}마리")
```

```python
samp1 = Bacteria(200, 2)
samp2 = Bacteria(400, 3)
samp3 = Bacteria(1400, 1.5)
```

```python
samp1.grow(10)
```

```python
samp3.grow(3)
```

```python
samp1.check()
samp3.check()
```

```python
# 예제4 캡슐화
class Bank:
    __interest_rate = 0.02 # 클래스 변수 p210
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance or 0
    def bprint(self): # 잔액조회
        return f"{self.owner }님의 잔고는 {self.__balance}원 입니다."
    def deposit(self, amount): # 입금
        self.__balance += amount
        return f"{self.owner }님의 계좌에 {amount}원이 입금 되었습니다. 현재잔액: {self.__balance}원"
    def withdraw(self, amount): # 출금
        if self.__balance >= amount:
            self.__balance -= amount
            return f"{self.owner }님의 계좌에서 {amount}원이 출금 되었습니다. 현재잔액: {self.__balance}원"
        return "잔액이 부족합니다."
    def interest(self):
        self.__balance = self.__balance * (1 + self.__interest_rate)
        return f"{self.__interest_rate}% 의 이자가 반영되어 현재잔액: {self.__balance:.2f}원"
```

```python
acc1 = Bank("철수", 5000)
acc2 = Bank("영희", 1000)
acc3 = Bank("유진스", 50000)
acc4 = Bank("병우스", 15000)
```

```python
print(acc1.bprint())
print(acc2.bprint())
print(acc3.bprint())
print(acc4.bprint())
```

```python
acc2.deposit(10000)
```

```python
acc4.withdraw(14000)
```

```python
# 예제5 응용
import random
class Character:
    def __init__(self, name, hp, attack_p):
        """케릭터 초기화"""
        self.name = name
        self.hp = hp
        self.attack_p = attack_p
    def attack(self, opponent):
        """상대방을 공격"""
        damage = random.randint(2, self.attack_p)
        opponent.hp -= damage
        print(f"{self.name}가 {opponent.name}을 공격하여 {damage}를 입혔습니다.")
    def now_status(self):
        """현재 체력 상태 확인"""
        return f"이름:{self.name} 체력:{self.hp}"
    def is_alive(self):
        """케릭터 생존여부 확인"""
        return self.hp > 0   
```

```python
# 히어로와 몬스터 생성
player = Character("영웅", attack_p = 15, hp = 50)
monster = Character("골렘", 65, 9)
```

```python
# 전투시작
import time as tt
print( f"{'전투시작':=^30}")
while player.is_alive() and monster.is_alive():
    # input("경기를 시작하려면 엔터를 치세요.")
    tt.sleep(2)
    if player.is_alive(): player.attack(monster)
    if monster.is_alive():monster.attack(player)
    print("투닥투닥","퍽퍽")
    print(player.now_status(), ' vs ', monster.now_status())    
# 결과
if player.is_alive():
    print("💐당신이 승리하셨습니다!!! 😍😍")
else:
    print("🩻패배하였습니다.. to continue..?")
```

## 상속과 메서드오버라이딩

```python
# 앞의 예제2번을 가져와서...
class Pet:
    def __init__(self, name, species):
        """펫을 생성"""
        self.name = name
        self.species = species
        self.hunger = 50 # 배고픔 수치 0 ~ 100    
    def feed(self, many = 10):
        """밥을 주면 배고픔 감소"""
        if self.hunger > 10:
            self.hunger -= many
        return f"{self.name}에게 밥을 줬어요. 배고픔:{self.hunger}"
    def play(self, how = 20):
        """놀아주면 배고픔 증가"""
        if self.hunger < 80:
            self.hunger += how
        return f"{self.name}와 잘 놀았어요~!! 배고픔:{self.hunger}"
```

```python
# 상속받아서 새로운 클래스 생성 + 메서드오버라이딩
class MyPet(Pet):
    def cry(self): # 상속 후 기능추가
        print("크르릉")
    def play(self): # 메서드 오버라이딩
        print("놀기싫다. 격렬하게 놀기싫다. 주인아..")
```

```python
dog5 = MyPet("깨갱","시고르잡종")
dog5.feed()
```
