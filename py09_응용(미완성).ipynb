```python
# 이터레이터 만들기

class MyIterator:
    def __init__(self, data):
        self.data = data
        self.position = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.position >= len(self.data):
            raise StopIteration
        result = self.data[self.position]
        self.position += 1
        return result
```

```python
it = MyIterator([1,2,3])
```

```python
for n in it:
    print(n)
```

```python
# 제너레이터
def mybot():
    try:
        print("작업시작")
        yield "1번작업완료"
        print("로봇A 위치로 동작")
        yield "2번작업완료"
        print("로봇A 그리퍼 동작후 위치반환")
        yield "3번작업완료"
        print("로봇B 가동시작")
        yield "end"
    except:
        print("더 이상 할 작업이 없습니다.")

gb = mybot()
```
