```python
# 1 클래스 상속받고 메서드 추가하기 1

class Calculator:
    def __init__(self):
        self. value = 0

    def add(self, val):
        self.value += val

class UpgradeCalculator(Calculator):
    def minus(self, val):
        self.value -= val

cal = UpgradeCalculator()
cal.add(10)
cal.minus(7)

print(cal.value)
```
