```python
# 로또 번호

# 1 ~ 45 임의의 수 중에서 중복없는 6개의 수를 뽑아 나열하여 분석하십시오.
# [44, 22, 23, 41, 42, 45]

# while 반복해서 1000이상으로 돌려서

#가장 많이 출현한 숫자를 뽑으세요.

import random

lotto_numbers = random.sample(range(1, 46), 6)

lotto_numbers.sort()

print(f"이번 주 추천 번호: {lotto_numbers}")


import random

counts = [0] * 46
iteration = 0

while iteration < 1000:
    lotto = random.sample(range(1,46),6)
    for num in lotto:
        counts[num] += 1

    iteration += 1

max_count = max(counts)
most_frequent_num = counts.index(max_count)

print(f"총 {iteration}번 시물레이션 완료!")
print(f"가장 많이 출연한 숫자: {most_frequent_num}번 (총 {max_count}회)")
```

```python
a = [int(((r*3)/2)*1.5) for r in range(0,100)]
b = [] # [짝수만]
c = [] # [홀수만]

for x in a:
    if x%2 ==0:
        b.append(x)
    else:
        c.append(x)

print(a, len(a))
print(b, len(b))
print(c, len(c))
```
