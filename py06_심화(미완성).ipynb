```python
# 만약 x100개 이상이라면??? 아주 불편한 코드

result1 = 0
result2 = 0

def add1(num):
    global result1
    result1 += num  # 결괏값(result)에 입력값(num) 더하기
    return result1  # 결괏값 리턴

def add2(num):
    global result2
    result2 += num  # 결괏값(result)에 입력값(num) 더하기
    return result2  # 결괏값 리턴
```

```python
add1(3)
```

```python
# 클래스 사용

class Cal:
    def __init__(self):
        self.result = 0
    
    def add(self, num):
        self.result += num
        return self.result

    def mul(self, num):
        self.result *= num
        return self.result
```

```python
cal1 = Cal()
cal2 = Cal()
```

```python
print(cal1.add(3))
```

```python
print(cal1.mul(3))
```

```python
# 클래스 구조 만들기
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
    def prints(self):
        print(self.first, self.second)
```

```python
four1 = FourCal()
```

```python
four1.setdata(3,4)
```

```python
four1.prints()
```

```python
# 예제1

class DDuk:
    def __inint__(self, won, ea):
        self.won = won
        self.ea = ea

    def dduk(self):
        result = self.won * self.ea
        print(f"배가 고파서 떡볶이집에서 떡볶이를 {result}원어치 먹었다.")

    def odeng(self):
        result = self.won * 2 * self.ea
        print(f"배가 고파서 떡볶이집에서 떡볶이를 {result}원어치 먹었다.")
```

```python
재형 = DDuk(2500,4)
호성 = DDuk(1500,7)
재훈 = DDuk(2000,5)
```

```python
재형.dduk()
재형.odeng()
```
